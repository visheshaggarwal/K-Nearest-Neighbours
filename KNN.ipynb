```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X,Y = make_classification(n_samples=400,n_features=2,n_classes=2,n_clusters_per_class=1,n_informative=2,n_redundant=0)
print(X.shape,Y.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
```

```python
def distance(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def Knn(X,Y,Query,k = 5):
    m = X.shape[0]
    dist = []
    
    for i in range(m):
        dist.append((distance(X[i],Query),Y[i]))
    
    dist.sort()
    dist = dist[:k]
    print(dist)
    class1 = 0
    class2 = 0
    for i in range(k):
        if dist[i][1] == 1:
            class2 += 1
        else:
            class1 +=1
            
    if class1>class2:
        return 0
    else:
        return 1
    
```

```python
ans = Knn(X,Y,[0,-1])
print(ans)
```
